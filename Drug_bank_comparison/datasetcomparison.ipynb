```python
import datacompy, pandas as pd 
```

```python
drug_Bank_Dataframe = pd.read_csv('drugbank_smiles.csv')
```

```python
drug_Bank_Dataframe.head()
```

```python
bindingdb_Dataframe = pd.read_csv('bindingdbtrain.csv')
```

```python
bindingdb_Dataframe.head()
```

# Comparing the DrugBank and BindingDB dataset

```python
compare_DrugBank_and_BindingDB = datacompy.Compare(
drug_Bank_Dataframe,
bindingdb_Dataframe,
join_columns='smiles',  
abs_tol=0, 
rel_tol=0, 
df1_name='DrugBank', #Optional, defaults to 'df1'
df2_name='BindingDB' #Optional, defaults to 'df2'
)
print(compare_DrugBank_and_BindingDB.report())
```
